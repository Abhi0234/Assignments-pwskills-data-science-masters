```python
Q1. Explain why we have to use the Exception class while creating a Custom Exception.
      Note: Here Exception class refers to the base class for all the exceptions.

Ans: In Python, the Exception class is the base class for all built-in exceptions, and it provides a set of common methods and 
     properties that can be used for error handling.
    
    When creating custom exceptions, it's recommended to derive them from the Exception class (or one of its subclasses) to ensure
    that they can be handled using the common error handling mechanisms provided by the language. This is because the built-in error
    handling mechanisms in Python, such as try-except blocks, are designed to work with exceptions that inherit from the Exception class.
    
    Deriving a custom exception class from Exception makes it possible to use the except clause to catch the custom exception, along
    with other built-in exceptions. This allows us to write more general-purpose exception handling code that can handle a range of
    exception types.
    
    Therefore, by deriving our custom exception class from the Exception class, we can ensure that it's compatible with the standard
    error handling mechanisms in Python, and that it includes the common error handling features that are used throughout the language.
```

```python
# Q4. Why LookupError class is used? Explain with an example KeyError and IndexError.

# Ans : LookupError: LookupError is a subclass of "Exception" class and base class of saveral built-classes including 
#                 "keyError" and "IndexError". It is used as the base class for exception that occur when a key or index
#                  is not found in collecion or mapping object. by using lookuperror class we can catch both types of errors
#                    using a single except block.
        
#     1. KeyError: It is the subclass of LookupError class and it raise an exception when trying to non-existing key in a dictionary
        
#         for example: In this example In the example above, we first try to access the key "pear" in the dictionary my_dict.
#                     Since "pear" is not a key in the dictionary, a KeyError is raised. We catch the exception using a try-except
#                     block and print the error message.
            
my_dict = {"apple": 1, "banana": 2, "orange": 3}
try:
    value = my_dict["pear"]
except KeyError as e:
    print(f"Error key not found : {e}")

#          2. Indexerror: It is the subclass of LookupError class and it raise an exception when trying to acces non-existing index in a
#                        list or other sequnce objct
         
#         for example:  we try to access the element at index 3 in the list my_list. Since my_list only has three elements 
#                       (with indices 0, 1, and 2), trying to access index 3 will raise an IndexError. We catch the exception  
#                       using a try-except block and print the error message.
            
my_list = [1, 2, 3]
try:
    value = my_list[3]
except IndexError as e:
    print(f"Error: {e}")
```

```python
Q6. List down some best practices for exception handling in python.

Ans: the some best practices for exception handling in python are:
    
    1 use specific exception : Always use specific class instead of using base class. This helps us to identify execr error and handle
                             it effectively.
        
    2 use logging : use logging module to log exception. which helps us to diagnose the issue.
     
    3 Use finally block : Always use the finally block to perform cleanup operations such as closing files, database connections, etc.,
                    regardless of whether an exception was thrown or not.
    
    4 don't ignore exception : don't ignore exception. Always handle exceptions and make sure your code continues to work correctly, 
                    even in exceptional circumstances.
    
    5 keep exception messages simple : Keep the error messages simple and clear, and avoid including sensitive information such as
                    passwords or other confidential information.
    
    6 Test your exception handling code: Test your exception handling code thoroughly to ensure it works as expected and to identify any
                    potential issues before your code goes into production.
```
