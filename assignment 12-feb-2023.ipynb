```python
# Q.3: Which python statement is used to handle exception in python? Explain with an Example.

# Ans: the "try" and "Except" statement is used to handle exception in python. where, try clause contains the block
#      of code that can raise an exception, while the except block contain the lines of code that handle the exception. 
    
#     for Example: In this Example we write a block of code in try clause which open a file "new.txt" in read mode. if 
                   # file "new.txt" dosen't exist then try clause throw an exception which is handled by Except clause.
    
    
try:        
    with open("new.txt", "r") as f:
        f.read()
except FileNotFoundError as e:
    print(f"File not found :", e)
```

```python
# Q.4 Explain with an Example: 
#     1 try and except
#     2 finally
#     3 raise

# Ans

# 1 try and except : the "try" and "Except" statement is used to handle exception in python. where, try clause contains 
#                    the blockof code that can raise an exception, while the except block contain the lines of code that
#                     handle the exception. 
    
#     for Example: In this Example we write a block of code in try clause which open a file "new.txt" in read mode. if 
                   # file "new.txt" dosen't exist then try clause throw an exception which is handled by Except clause.
        
try:        
    with open("new.txt", "r") as f:
        f.read()
except Exception as e:
    print(e)

    
# 2 finally : In python, The finally statement is used in exception handling. the code of finally will be executed after the 
#             block of code inside try and except have been excuted. the code of finally will be executed regardless wheather
#             an exception is raised or not
        
#             The finally is used to release the connections like : open file, database connection, etc. it is also ensure that
#             some specific code will be executed, wheather the extion an exception is raised or not.
    
#     for Example: In this example 5 is divided by 0 which gives ZeroDivisionError but still the code in finally block is executed
try:
    k = 5/0 # raises divide by zero exception.
    print(k)
  

except ZeroDivisionError:   
    print("Can't divide by zero")
      
finally:
    # this block is always executed 
    # regardless of exception generation.
    print('This is always executed') 


# # 3 raise : The raise statement is used to throw an exception or error and stops the program at any time if the certain condition
#              doesn't meet. When we use raise statement. there is no compulsion to give an exception class along with it. 
        
#              Raise allows us to throw one exception at any time. It helps us raise exceptions when we may run into situations where
#              execution canâ€™t proceed. It also helps us in reraise exception that is caught.
    
#       for Example: In this example the conditions doesn't match and raise statement throws an exception in try clause, then by try 
#                     clause the exception is handled by the except clause.
        
x = -1
try:
    if x < 0:
      raise Exception("Sorry, no numbers below zero")
except Exception as e:
    print(e)
    
```

```python
# Q.5 : What are custom exception in python? why do we need custom exception? Explain with an example.

# Ans: In python, a custom exception is a user- defined exception, that can be raised like build in exception.
#      It allows us to define their own exception which is useful in specific situations.

#         It is useful to create custom exception when the built-in exception do not provide enough information or
#         context about the error.
        
#         the custom exception class is created by inherating the built-in "Exception" class or any of its derived classes
#         in the base class. once a custom exception class has been defined. it can be raised using "raise" statement.    

        
#     for Example:
#             In this example, we have defined a custom exception NegativeNumberError that inherits from the Exception class.
#             This custom exception takes a value argument and provides a custom error message when a negative number is encountered.

#               We have also defined a function square_root that calculates the square root of a given number. If the input number is
#               negative, the NegativeNumberError is raised, which will be caught by the except block.

#               When we call the square_root function with a negative number, it raises the NegativeNumberError, which is caught by the
#             except block. The error message " -1 is a negative number" is printed to the console.

class NegativeNumberError(Exception):
    def __init__(self, value):
        self.value = value

    def __str__(self):
        return f"{self.value} is a negative number"


def square_root(num):
    if num < 0:
        raise NegativeNumberError(num)
    else:
        return num ** 0.5


try:
    result = square_root(-1)
    print(result)
except NegativeNumberError as e:
    print(e)
    
```
